```python
### CALCULATES THE AVERAGE DIFFERENCES BETWEEN THE ASTROMETRIC COORDINATES OF OUR SOURCES AND THEIR RESPECTIVE GAIA COUNTERPARTS ###

from astropy.io import ascii
from astropy.table import unique
import os
%run Crossmatch.ipynb

# The Gaia EDR3 catalog
gaia = ascii.read('/home/yobd/Documents/CatalogsCrA/New_Run/Gaia_DR3.dat')

# The list of mean difference in the right ascenscion between each detector and Gaia
ra_mean = []

# The list of standard deviations in the difference in the right ascenscion between each detector and Gaia
std_ra = []

# The list of mean difference in the declination between each detector and Gaia
dec_mean = []

# The list of standard deviations in the difference in the declination between each detector and Gaia
std_dec = []

# The list with the number of sources cross-matched with Gaia in each detector
n = []

# The path where the dataset is located
path = '/home/yobd/Documents/AstroAliSciObj/W-S-I+/MINAREA_4_THRESH_2_dot_7_SNR_20_SATUR_LEVEL_B4_PSFEX_30000_SMP_VAR_0_dot_2_SEEING_6_dot_7_7_MAG_PSF/Catalogs_Finished/'

# Encodes the path of the dataset to be used by other OS methods
folder = os.fsencode(path)

for item in os.listdir(folder):
    
    # Decodes the current item so it becomes a string
    item = os.fsdecode(item)
    
    # To comply with the func() function parameter
    item = path + item
    
    # Reads the data of the current detector
    wsi = ascii.read(item)
    
    # Cross-matches the current detector with the Gaia EDR3 catalog
    wsi_gaia = cross_match(wsi, 'ra', 'dec', gaia, 'RA_ICRS', 'DE_ICRS', 1, sep = True)
    
    # Sorts the table by the respective distance between cross-matches
    wsi_gaia.sort('d2d')
    
    # Removes the duplicate entries from the cross-matched catalog
    wsi_gaia = unique(wsi_gaia, keys = ['RA_ICRS', 'DE_ICRS'], keep = 'first', silent = True)
    
    # Calculates the mean difference in the right-ascenscion between the sources in this chip and their values in Gaia EDR3
    ra_mean_temp = np.nanmean(wsi_gaia['ra'] - wsi_gaia['RA_ICRS'])
    # Appends this value to the list of mean differences in the right-ascenscion
    ra_mean.append(ra_mean_temp)
    
    # Calculates the standard deviation of the mean difference in the right-ascenscion between the sources in this chip and their values in Gaia EDR3
    ra_std_temp = np.nanstd(wsi_gaia['ra'] - wsi_gaia['RA_ICRS'])
    # Appends this value to the list of standard deviations in the right-ascenscion
    std_ra.append(ra_std_temp)
    
    # Calculates the mean difference in the declination between the sources in this chip and their values in Gaia EDR3
    dec_mean_temp = np.nanmean(wsi_gaia['dec'] - wsi_gaia['DE_ICRS'])
    # Appends this value to the list of mean differences in the declination
    dec_mean.append(dec_mean_temp)
    
    # Calculates the standard deviation of the mean difference in the declination between the sources in this chip and their values in Gaia EDR3
    dec_std_temp = np.nanstd(wsi_gaia['dec'] - wsi_gaia['DE_ICRS'])
    # Appends this value to the list of standard deviations in the declination
    std_dec.append(dec_std_temp)
    
    # Appends the number of cross-matched sources to the respective list
    n.append(len(wsi_gaia))

# Calculates the mean of the means
ra_mean_diff = np.nanmean(np.array(ra_mean))

# The numerator in the equation for the overall standard deviation
std_numerator = 0

# The iterator to be used inside the cycle
i = 0

# Cycles through each one of the ten chips
while i < len(n):
    
    # Updates the numerator value
    std_numerator = std_numerator + (n[i] - 1) * std_ra[i] * std_ra[i] + n[i] * (ra_mean[i] - ra_mean_diff) * (ra_mean[i] - ra_mean_diff)
    
    # Iterates
    i += 1

# The denominator in the equation for the overall standard deviation
std_denominator = sum(n) - 1

# The overall standard deviation in the right-ascenscion
ra_std_diff = np.sqrt(std_numerator / std_denominator)

# Calculates the mean of the means
dec_mean_diff = np.nanmean(np.array(dec_mean))

# The numerator in the equation for the overall standard deviation
std_numerator = 0

# The iterator to be used inside the cycle
i = 0

# Cycles through each one of the ten chips
while i < len(n):
    
    # Updates the numerator value
    std_numerator = std_numerator + (n[i] - 1) * std_dec[i] * std_dec[i] + n[i] * (dec_mean[i] - dec_mean_diff) * (dec_mean[i] - dec_mean_diff)
    
    # Iterates
    i += 1

# The denominator in the equation for the overall standard deviation
std_denominator = sum(n) - 1

# The overall standard deviation in the declination
dec_std_diff = np.sqrt(std_numerator / std_denominator)

# Converts into arc-seconds
ra_mean_diff = 3600 * ra_mean_diff
dec_mean_diff = 3600 * dec_mean_diff
ra_std_diff = 3600 * ra_std_diff
dec_std_diff = 3600 * dec_std_diff

print("The mean right-ascenscion measurements difference between our W-S-I+ catalog and Gaia EDR3 is:", ra_mean_diff, "+/-", ra_std_diff, "arc-seconds.")

print("The mean declination measurements difference between our W-S-I+ catalog and Gaia EDR3 is:", dec_mean_diff, "+/-", dec_std_diff, "arc-seconds.")

### FOR THE N-A-L656 FILTER ###

# The list of mean difference in the right ascenscion between each detector and Gaia
ra_mean = []

# The list of standard deviations in the difference in the right ascenscion between each detector and Gaia
std_ra = []

# The list of mean difference in the declination between each detector and Gaia
dec_mean = []

# The list of standard deviations in the difference in the declination between each detector and Gaia
std_dec = []

# The list with the number of sources cross-matched with Gaia in each detector
n = []

# The path where the dataset is located
path = '/home/yobd/Documents/AstroAliSciObj/N-A-L656/MINAREA_4_THRESH_2_dot_7_SNR_20_SATUR_LEVEL_B4_PSFEX_30000_SMP_VAR_0_dot_2_SEEING_6_dot_7_7_MAG_PSF/Catalogs_Finished/'

# Encodes the path of the dataset to be used by other OS methods
folder = os.fsencode(path)

for item in os.listdir(folder):
    
    # Decodes the current item so it becomes a string
    item = os.fsdecode(item)
    
    # To comply with the func() function parameter
    item = path + item
    
    # Reads the data of the current detector
    h_alpha = ascii.read(item)
    
    # Cross-matches the current detector with the Gaia EDR3 catalog
    h_alpha_gaia = cross_match(h_alpha, 'ra', 'dec', gaia, 'RA_ICRS', 'DE_ICRS', 1, sep = True)
    
    # Sorts the table by the respective distance between cross-matches
    h_alpha_gaia.sort('d2d')
    
    # Removes the duplicate entries from the cross-matched catalog
    h_alpha_gaia = unique(h_alpha_gaia, keys = ['RA_ICRS', 'DE_ICRS'], keep = 'first', silent = True)
    
    # Calculates the mean difference in the right-ascenscion between the sources in this chip and their values in Gaia EDR3
    ra_mean_temp = np.nanmean(h_alpha_gaia['ra'] - h_alpha_gaia['RA_ICRS'])
    # Appends this value to the list of mean differences in the right-ascenscion
    ra_mean.append(ra_mean_temp)
    
    # Calculates the standard deviation of the mean difference in the right-ascenscion between the sources in this chip and their values in Gaia EDR3
    ra_std_temp = np.nanstd(h_alpha_gaia['ra'] - h_alpha_gaia['RA_ICRS'])
    # Appends this value to the list of standard deviations in the right-ascenscion
    std_ra.append(ra_std_temp)
    
    # Calculates the mean difference in the declination between the sources in this chip and their values in Gaia EDR3
    dec_mean_temp = np.nanmean(h_alpha_gaia['dec'] - h_alpha_gaia['DE_ICRS'])
    # Appends this value to the list of mean differences in the declination
    dec_mean.append(dec_mean_temp)
    
    # Calculates the standard deviation of the mean difference in the declination between the sources in this chip and their values in Gaia EDR3
    dec_std_temp = np.nanstd(h_alpha_gaia['dec'] - h_alpha_gaia['DE_ICRS'])
    # Appends this value to the list of standard deviations in the declination
    std_dec.append(dec_std_temp)
    
    # Appends the number of cross-matched sources to the respective list
    n.append(len(h_alpha_gaia))

# Calculates the mean of the means
ra_mean_diff = np.nanmean(np.array(ra_mean))

# The numerator in the equation for the overall standard deviation
std_numerator = 0

# The iterator to be used inside the cycle
i = 0

# Cycles through each one of the ten chips
while i < len(n):
    
    # Updates the numerator value
    std_numerator = std_numerator + (n[i] - 1) * std_ra[i] * std_ra[i] + n[i] * (ra_mean[i] - ra_mean_diff) * (ra_mean[i] - ra_mean_diff)
    
    # Iterates
    i += 1

# The denominator in the equation for the overall standard deviation
std_denominator = sum(n) - 1

# The overall standard deviation in the right-ascenscion
ra_std_diff = np.sqrt(std_numerator / std_denominator)

# Calculates the mean of the means
dec_mean_diff = np.nanmean(np.array(dec_mean))

# The numerator in the equation for the overall standard deviation
std_numerator = 0

# The iterator to be used inside the cycle
i = 0

# Cycles through each one of the ten chips
while i < len(n):
    
    # Updates the numerator value
    std_numerator = std_numerator + (n[i] - 1) * std_dec[i] * std_dec[i] + n[i] * (dec_mean[i] - dec_mean_diff) * (dec_mean[i] - dec_mean_diff)
    
    # Iterates
    i += 1

# The denominator in the equation for the overall standard deviation
std_denominator = sum(n) - 1

# The overall standard deviation in the declination
dec_std_diff = np.sqrt(std_numerator / std_denominator)

# Converts into arc-seconds
ra_mean_diff = 3600 * ra_mean_diff
dec_mean_diff = 3600 * dec_mean_diff
ra_std_diff = 3600 * ra_std_diff
dec_std_diff = 3600 * dec_std_diff

print("The mean right-ascenscion measurements difference between our N-A-L656 catalog and Gaia EDR3 is:", ra_mean_diff, "+/-", ra_std_diff, "arc-seconds.")

print("The mean declination measurements difference between our N-A-L656 catalog and Gaia EDR3 is:", dec_mean_diff, "+/-", dec_std_diff, "arc-seconds.")
```
